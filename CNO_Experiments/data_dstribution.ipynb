```python
import numpy as np
import matplotlib.pyplot as plt

# Load your dataset
# Replace this with your actual loading logic
data = np.load('/data/group_data/sage_lab_complex_geometry/FPO_Cylinder_Multiple_Hole_HighRe.npy')  # shape (N, T, H, W, C)

# Extract Reynolds number from the 4th channel (index 3)
reynolds_data = data[..., 3]  # shape: (N, T, H, W)

# Flatten the data to a 1D array for histogram plotting
reynolds_flat = reynolds_data.flatten()

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(reynolds_flat, bins=100, color='skyblue', edgecolor='black')
plt.title('Distribution of Reynolds Number')
plt.xlabel('Reynolds Number')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```
